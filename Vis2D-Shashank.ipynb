```python
import pickle

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.ticker import NullFormatter
from sklearn.manifold import TSNE, Isomap
from sklearn.decomposition import PCA
```

```python
import umap
```

```python
### Load SoftMax
model_name = "resnet"
train_name = "IN"
test_name = "IN"

classes = [
    'airplane', 'bear', 'bicycle', 'bird', 
    'boat', 'bottle', 'car', 'cat', 
    'chair', 'clock', 'dog', 'elephant',
    'keyboard', 'knife', 'oven', 'truck'
]

markers = ["^", "s", "p", "X", "D"]

save_figpath = './vis2d_sample.png'

m_size = 200
moreplot = 10
num_samples = 400

cmap = "tab10"

load_path = f'./softmax_dict/{test_name}_to_{model_name}-{train_name}_softmax_dict.pkl'
save_path = f'./manifold_dict/{test_name}_to_{model_name}-{train_name}_softmax'
```

```python
with open(load_path, mode='rb') as f:
    dicts = pickle.load(f)
```

```python
X = np.empty((0, 1000))
y = np.empty((0, ))

for i, c in enumerate(classes):
    X = np.append(X, dicts[c], axis=0).astype(float)
    y = np.append(y, [i]*dicts[c].shape[0], axis=0).astype(int)

# test labels
y = y[:num_samples]
```

```python
# test data param
test_index = np.array([1,81,161,241,321])
print(y[test_index])
```

```python
### Visualizing 2D
fig = plt.figure(1, figsize=(20, 5))

manifolds_fns = [] 
manifolds_fns += [[TSNE(n_components=2, random_state=0, perplexity=30, n_iter=1000, n_jobs=-1), 't-SNE']]
manifolds_fns += [[Isomap(n_neighbors=10, n_components=2, n_jobs=-1), 'IsoMap']]
manifolds_fns += [[PCA(n_components=2), 'PCA']]
manifolds_fns += [[umap.UMAP(), 'UMAP']]

n = len(manifolds_fns)
for idx in range(n):

    manifold_fn   = manifolds_fns[idx][0]
    manifold_name = manifolds_fns[idx][1]
    
    fname = f"{save_path}_{manifold_name}.npy"         
    try:
        X_embedded = np.load(fname)
    except:
        X_embedded = manifold_fn.fit_transform(X)
        np.save(fname, np.array(X_embedded))
    
    X_embedded = X_embedded[:num_samples]
    
    ax = fig.add_subplot(1,n, idx+1)

    for train_num in range(5):
        index = train_num*80
        ax.scatter(
                    X_embedded[index:index+moreplot, 0], X_embedded[index:index+moreplot, 1],
                    c=y[index:index+moreplot], cmap=cmap,
                    vmin=0, vmax=4, s=50, alpha=0.8
        )

    m_counter = 0
    for x1, x2 in zip(X_embedded[test_index][:, 0], X_embedded[test_index][:, 1]):
        ax.scatter(
            x1, x2,
            c="black",
            s=m_size, alpha=0.9,
            marker=markers[m_counter]
        )
        m_counter += 1

    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    ax.set_title(manifold_name)

fig.patch.set_alpha(0)
fig.savefig(save_figpath)
```
